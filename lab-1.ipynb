```python
import numpy as np
import pandas as pd

pd.set_option('display.max_rows', None)
pd.set_option('display.max_columns', None)
pd.set_option('display.max_colwidth', None)
pd.set_option('display.float_format', '{:.4f}'.format)
```

```python
with open("datasets\\brooklyn_sales_map.csv") as f:
    ncols = len(f.readline().split(','))

sales_data = pd.read_csv(filepath_or_buffer="datasets\\brooklyn_sales_map.csv",
                            usecols=range(1,ncols),
                            index_col="zip_code",
                            low_memory=False)

sales_data.head(10)
```

```python
sales_data["age"] = pd.Series(np.random.randint(low=0, high=100, size=len(sales_data.index)),
                                index=sales_data.index)
sales_data.head(10)
```

```python
mean_age = sales_data['age'].mean()
umean_age_count = len(sales_data[sales_data.age > mean_age].index)

print(f"Средний возраст построек: {mean_age.round(2)}\n"+
        f"Количество построек, возраст которых больше среднего: {umean_age_count}")
```

```python
column_nan_percentge = (sales_data.isna().sum(axis=0) / len(sales_data.index) * 100).round(2)
column_nan_percentge
```

```python
sales_data_slice = pd.DataFrame(data=sales_data,
                                columns=['borough', 'neighborhood', 'building_class_category',
                                            'tax_class', 'building_class', 'gross_sqft',
                                             'total_units', 'age', 'sale_price'])

print(f"Количество строк до удаления пропущенных значений: {len(sales_data_slice.index)}")
```

```python
sales_data_slice.dropna(how='all')

print(f"Количество строк после удаления пропущенных значений: {len(sales_data_slice.index)}")
```

```python
for i in range(0,10):
    sales_data_slice = sales_data_slice.append(pd.Series(dtype='object'), ignore_index=True)

sales_data_slice.tail(15)
```

```python
sales_data_slice.dropna(how='all', axis=0, inplace=True)

print(f"Количество строк после повторного удаления пропущенных значений: {len(sales_data_slice.index)}")
```
